```python
# Lista de columnas a procesar
columnas_a_procesar = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count']

# Aplicar el reemplazo en cada columna
for columna in columnas_a_procesar:
    vg[columna] = vg[columna].astype(str).str.replace('.', ',')

# Exportar el DataFrame a Tableau (o a otro formato si prefieres)
vg.to_csv('../data/vg_cleaned.csv', index=False)
```

```python
# Reemplazar punto por coma en las columnas numéricas
vg['Global_Sales'] = vg['Global_Sales'].astype(str).str.replace('.', ',')

# Exportar el DataFrame a Tableau
vg.to_csv('../data/vg_cleaned.csv', index=False)
```
